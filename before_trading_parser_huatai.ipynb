```python
from urllib import request, parse
import numpy as np
import pandas as pd
import time
```

```python
hs_origin = list(range(1132,1418+1))
ss_origin = list(range(1,261+1)) + list(range(1090,1241+1))
hs_df = pd.DataFrame()
ss_df = pd.DataFrame()
todate = '2019-08-05'
```

```python
while(True):
    hs_new = []
    ss_new = []
    for hs_page_num in hs_origin:
        data = parse.urlencode([('date', todate),('hsPage', str(hs_page_num)),('hsPageSize', '5'),('ssPage', '1'),('ssPageSize', '5'),])
        req = request.Request('http://web.htsc.com.cn/browser/rzrqPool/getDbZqc.do')
        req.add_header('Cookie', 'JSESSIONID=86C0EEE60666B49376D1A24668017D5C; route=b3ae896a9ce5631e701a8603f6562842; _pk_id.2.b610=35b6117c6c140f72.1544504036.3.1550218520.1550040904.; _pk_ses.2.b610=*')   
        req.add_header('Host', 'web.htsc.com.cn')
        req.add_header('Origin', 'http://web.htsc.com.cn')
        req.add_header('Referer', 'http://web.htsc.com.cn/browser/rzrq/marginTrading/bdzqc/target_stock_pool_bail.jsp')
        req.add_header('User-Agent', 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36')
        req.add_header('X-Requested-With', 'XMLHttpRequest')
        try:
            with request.urlopen(req,data=data.encode('utf-8'),timeout=10) as f:
                print('Status:', f.status, f.reason)
                result = f.read().decode('utf-8')
            try:
                result = result.replace('"','').replace('}','').replace('{','').replace(']','').replace(':[',',').split(',')
                result = result[result.index('dbHs')+1:-1]
                if len(result)%21 != 0:
                    raise ValueError
                result = np.array(result).reshape(len(result)//21,21)
                for result_term in result:
                    hs_add = pd.DataFrame([[result_term[3].split(':')[-1],result_term[4].split(':')[-1],result_term[20].split(':')[-1]]])
                    hs_df = pd.concat([hs_df,hs_add])
                #hs_df.to_excel('华泰证券沪市.xlsx',index=False,encoding='gbk')
                print(hs_page_num)
            except:
                print('*****************',hs_page_num)
                hs_new.append(hs_page_num)
        except:
            print('*****************',hs_page_num)
            hs_new.append(hs_page_num)
        time.sleep(2)
    for ss_page_num in ss_origin:
        data = parse.urlencode([('date', todate),('hsPage', '1'),('hsPageSize', '5'),('ssPage', str(ss_page_num)),('ssPageSize', '5'),])
        req = request.Request('http://web.htsc.com.cn/browser/rzrqPool/getDbZqc.do')
        req.add_header('Cookie', 'JSESSIONID=86C0EEE60666B49376D1A24668017D5C; route=b3ae896a9ce5631e701a8603f6562842; _pk_id.2.b610=35b6117c6c140f72.1544504036.3.1550218520.1550040904.; _pk_ses.2.b610=*')   
        req.add_header('Host', 'web.htsc.com.cn')
        req.add_header('Origin', 'http://web.htsc.com.cn')
        req.add_header('Referer', 'http://web.htsc.com.cn/browser/rzrq/marginTrading/bdzqc/target_stock_pool_bail.jsp')
        req.add_header('User-Agent', 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36')
        req.add_header('X-Requested-With', 'XMLHttpRequest')
        try:
            with request.urlopen(req, data=data.encode('utf-8'), timeout=10) as f:
                print('Status:', f.status, f.reason)
                result = f.read().decode('utf-8')
            try:
                result = result.replace('"','').replace('}','').replace('{','').replace(']','').replace(':[',',').split(',')
                result = result[result.index('dbSs')+1:result.index('dbHs')-2]
                if len(result)%21 != 0:
                    raise ValueError
                result = np.array(result).reshape(len(result)//21,21)
                for result_term in result:
                    ss_add = pd.DataFrame([[result_term[3].split(':')[-1],result_term[4].split(':')[-1],result_term[20].split(':')[-1]]])
                    ss_df = pd.concat([ss_df,ss_add])
                #ss_df.to_excel('华泰证券深市.xlsx',index=False,encoding='gbk')
                print(ss_page_num)
            except:
                print('*****************',ss_page_num)
                ss_new.append(ss_page_num)
        except:
            print('*****************',ss_page_num)
            ss_new.append(ss_page_num)
        time.sleep(2)
    if hs_new == [] and ss_new == []:
        break
    else:
        hs_origin = hs_new
        ss_origin = ss_new
```

```python
df = pd.concat([hs_df,ss_df])
df.columns = ['证券代码','担保折算率','证券名称']
df = df.sort_values('证券代码')
df['证券代码'] = df['证券代码'].astype(int)
df['担保折算率'] = df['担保折算率'].astype(float)
df = df.loc[:,['证券代码','证券名称','担保折算率']]
df = df[(df['证券代码']<5000)|((df['证券代码']>300000)&(df['证券代码']<400000))|((df['证券代码']>=600000)&(df['证券代码']<700000))]
df.to_excel('华泰证券.xlsx',index = False)
```
